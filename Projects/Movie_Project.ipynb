```python
Rate = imdb_df["Rate"].value_counts()
numbers = Rate.index
quantity = Rate.values
import plotly.express as px
fig = px.pie(imdb_df, values=quantity, names=numbers)
fig.show()
```

```python
Metascore = imdb_df["Metascore"].value_counts()
numbers = Metascore.index
quantity =Metascore.values
import plotly.express as px
fig = px.pie(imdb_df, values=quantity, names=numbers)
fig.show()
```
